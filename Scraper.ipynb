```python
import re
from PyPDF2 import PdfReader
import pandas as pd
from IPython.display import HTML
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
#1. Scrape content
#2. Create pandas dataframe with clickable links and add notes + framework column
#3. Share work with Team


reader = PdfReader("ReviewPaper.pdf")
text = ""
for page in reader.pages:
    text += page.extract_text()
```

```python
classificationTasks = ""
regressionTasks = ""

text=text.replace('\n', ' ').replace('\r', '')
text = ''.join(text)

classificationTasks = re.search('aAbbreviations:(.*?).L. Zhao, Y. Zhu, J. Wang et al. Computational and Structural Biotechnology', text).group(1)+"$"

regressionTasks = re.search('network blocks Fully(.*?)2831–2838', text).group(1)

classificationTasks = re.search('the listed tools:(.+?)$', classificationTasks).group(1)[:-1]

classificationTasks = classificationTasks.split(";")

regressionTasks = re.search('tools:(.+?).L. Zhao', regressionTasks).group(1)

regressionTasks = regressionTasks.split(";")

dataC=[]
for i in classificationTasks:
    i = i.split("–")
    dataC.append(i)

# Create the pandas DataFrame
df = pd.DataFrame(dataC, columns=['Name', 'Link', "Framework"])

dataR=[]
for i in regressionTasks:
    i = i.split("–") + [" - None"]
    dataR.append(i)

dfR = pd.DataFrame(dataR, columns=['Name', 'Link', "Framework"])

HTML(df.to_html(render_links=True, escape=False))
```

```python
frameworks = ["Theano","Scikitlean","Pytorch","Tensorflow","Tensorflow","Pytorch" ,"Pytorch","","Pytorch","Pytorch","","Pytorch","Pytorch","Pytorch","Pytorch","","Pytorch",""]

notes = ["","", "Link instead https://github.com/masashitsubaki/CPI_prediction", "10 = https://github.com/enai4bio/BridgeDPI"] 

for i in range(len(frameworks)):
    df.iloc[i][2] = frameworks[i]
    if "CPI-GNN" in df.iloc[i][0]:
        df.iloc[i][1] = "https://github.com/masashitsubaki/CPI_prediction"
for i in range (len(df.index)):
    if df.loc[i][2] != "Pytorch":
        df=df.drop([i])
df.iloc[4][1] = "https://github.com/enai4bio/BridgeDPI"

HTML(df.to_html(render_links=True, escape=False))
```
