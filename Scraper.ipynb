```python
import re
from PyPDF2 import PdfReader
import pandas as pd
from IPython.display import HTML
```

```python
#1. Scrape content
#2. Create pandas dataframe with clickable links and add notes + framework column
#3. Share work with Team


reader = PdfReader("ReviewPaper.pdf")
text = ""
for page in reader.pages:
    text += page.extract_text() + "\n"
```

```python
#first get the relevant sections
classificationTasks = ""
regressionTasks = ""

text = ''.join(text)


classificationTasks = re.search('aAbbreviations:(.*?).L. Zhao, Y. Zhu, J. Wang et al. Computational and Structural Biotechnology', text).group(1)+"$"


regressionTasks = re.search('network blocksFully(.*?)2831–28382835supervised learning,', text).group(1)
#print(regressionTasks)

# 1 - getting all links with name 
#  -split at ; and -
#3. create table with 3 columns name, link, framework
#4. create table with all entries for first 2 columns last empty
#5.go through repo and add 3 column 


classificationTasks = re.search('the listed tools:(.+)$', classificationTasks).group(1)[:-1]

classificationTasks = classificationTasks.split(";")
name = []
link = []

data=[]
for i in classificationTasks:
    i = i.split("–")
    data.append(i)
        
# Create the pandas DataFrame
df = pd.DataFrame(data, columns=['Name', 'Link', "Framework"])
  
HTML(df.to_html(render_links=True, escape=False))
```
